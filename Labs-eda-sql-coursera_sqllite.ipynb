## Tareas

Ahora escriba y ejecute consultas SQL para resolver las tareas de asignación.

**Nota: si los nombres de las columnas están en mayúsculas y minúsculas, enciérrelo entre comillas dobles
Por ejemplo, "Landing_Outcome"**

### Tarea 1

##### Mostrar los nombres de los sitios de lanzamiento únicos en la misión espacial

```python
%sql select DISTINCT "Launch_Site" from SPACEXTBL;
```

```python
%%sql
SELECT COUNT(*) LAUNCH_SITE
FROM SPACEXTBL
WHERE LAUNCH_SITE = 'CCAFS LC-40';
```

```python
%%sql
SELECT COUNT(*) LAUNCH_SITE
FROM SPACEXTBL
WHERE LAUNCH_SITE = 'CCAFS SLC-40';
```

### Tarea 2

##### Mostrar 5 registros donde los sitios de lanzamiento comienzan con la cadena 'CCA'

```python
%%sql
SELECT * FROM SPACEXTBL
WHERE LAUNCH_SITE LIKE 'CCA%' LIMIT 5;
```

### Tarea 3

##### Muestra la masa de carga útil total transportada por los propulsores lanzados por la NASA (CRS)

```python
%%sql
SELECT sum("PAYLOAD_MASS__KG_") as Carga_total_trasportada from SPACEXTBL
where Customer = "NASA (CRS)";
```

### Tarea 4

##### Muestra la masa de carga útil promedio transportada por la versión de refuerzo F9 v1.1

```python
%%sql
select  avg("PAYLOAD_MASS__KG_") as Promedio_Carga_util from SPACEXTBL
where Booster_Version like "F9 v1.1%"
```

### Tarea 5
##### Indique la fecha en que se logró el primer aterrizaje exitoso en la plataforma de tierra.
*Sugerencia: use la función min*

%%sql min("Date") from SPACEXTBL
WHERE 

```python
%%sql
SELECT min(Date) FROM SPACEXTBL
where  Landing_Outcome  = 'Success (ground pad)' ;

```

### Tarea 6

##### Enumere los nombres de los propulsores que tienen éxito en naves no tripuladas y tienen una masa de carga superior a 4000 pero inferior a 6000

```python
%%sql
SELECT BOOSTER_VERSION
FROM SPACEXTBL
WHERE LANDING_OUTCOME = 'Success (drone ship)'
    AND 4000 < PAYLOAD_MASS__KG_ < 6000 ;
```

### Tarea 7

##### Indique el número total de resultados de misiones exitosas y fallidas

```python
%%sql 
select count("Mission_Outcome") as Exitoso from SPACEXTBL
where Mission_Outcome like "Success%";
```

```python
%%sql 
select count("Mission_Outcome") - 100 as Fallidos from SPACEXTBL;
```

```python
%%sql
SELECT MISSION_OUTCOME, COUNT(MISSION_OUTCOME) AS TOTAL_NUMBER
FROM SPACEXTBL
GROUP BY MISSION_OUTCOME;
```

### Tarea 8

##### Enumere los nombres de las booster_versions que han transportado la masa máxima de carga útil. Usar una subconsulta

```python
%%sql 
select distinct(Booster_Version) from SPACEXTBL
where PAYLOAD_MASS__KG_ = (
    select max(PAYLOAD_MASS__KG_)
    from SPACEXTBL);
```

### Tarea 9

##### Enumere los registros que mostrarán los nombres de los meses, los resultados de los aterrizajes fallidos en el barco de drones, las versiones de refuerzo, el sitio de lanzamiento para los meses del año 2015.

**Nota: SQLLite no admite nombres de meses. Entonces necesita usar substr(Date, 4, 2) como mes para obtener los meses y substr(Date,7,4)='2015' para el año.**

```python
%%sql
SELECT LANDING_OUTCOME, BOOSTER_VERSION, LAUNCH_SITE
FROM SPACEXTBL
WHERE Landing_Outcome = 'Failure (drone ship)'
    AND YEAR(DATE) = '2015';
```

### Tarea 10

##### Clasifique el recuento de resultados de aterrizaje exitosos entre la fecha 04-06-2010 y 20-03-2017 en orden descendente.

```python
%%sql
SELECT LANDING_OUTCOME, COUNT(LANDING_OUTCOME) AS TOTAL_NUMBER
FROM SPACEXTBL
WHERE DATE BETWEEN '04-06-2010' AND '20-03-2017'
GROUP BY LANDING_OUTCOME
ORDER BY TOTAL_NUMBER DESC
```
